```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
from skimage import io
```

```python
husky = io.imread("https://img.freepik.com/premium-photo/dog-walks-mountains_477814-6118.jpg?w=1380")
plt.imshow(husky)
plt.title("Original")
plt.show()
```

```python
husky_gray = cv2.cvtColor(husky,cv2.COLOR_BGR2GRAY)
```

```python

husky_power_law=np.ones((husky_gray.shape[0],husky_gray.shape[1]))
GAMMA=3.5
for i in range(husky_gray.shape[0]):
    for j in range(husky_gray.shape[1]):
            husky_power_law[i][j]=(husky_gray[i][j]/255)**(GAMMA)
```

```python
def plot_image(image_1, image_2, title_1="Gray Husky", title_2="Power Lawed Husky"):
    plt.figure(figsize=(17,17))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.show()
```

```python
plot_image(husky_gray,husky_power_law)
```
